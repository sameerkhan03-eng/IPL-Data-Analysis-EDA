## --> Remove Duplicates & Remove NAN rows 

```python
## --> Remove Duplicates
df_matches.drop_duplicates(inplace=True)
df_players.drop_duplicates(inplace=True)
df_bowling.drop_duplicates(inplace=True)
df_deliveries.drop_duplicates(inplace=True)
df_players.dropna(how='all',inplace=True)
df_bowling.dropna(how='all',inplace=True)
```

## --> Fix Data Types

```python
## Fix Data types
#print(df_matches.info())
df_matches['date']=pd.to_datetime(df_matches['date'])

print(df_players.info())
cols=['match_id','runs','balls','fours','sixes']
df_players[cols]=df_players[cols].astype(int)
df_players

print(df_bowling.info())
col=['match_id','maidens','runs','wickets']
df_bowling[col]=df_bowling[col].astype(int)

df_deliveries['runs']=df_deliveries['runs'].astype(int)
print(df_bowling)
df_bowling.rename(columns={'runs_conceded':'runs'},inplace=True)
df_bowling.info()
```

## 7) Feature Engineering 

#### 7.1) Strike_rate (Recalculate)

```python
df_players['strike_rate']=(df_players['runs'] / df_players['balls'])*100 
```

#### 7.2) bowling Economy 


```python
#7.2) bowling Economy 
df_bowling['economy']=df_bowling['runs'] / df_bowling['overs'] 
#df_bowling['economy']=df_bowling['economy'].round(2)
#
df_bowling
```

#### 7.3) Total Runs Per Ball

```python
# Total Runs per ball
df_deliveries['total_runs']=df_deliveries['runs'] + df_deliveries['extras']
```

#### 7.4) Match Phase Creation 

```python
#### 7.4) Match Phase Creation 
def get_phase(over):
    if over <=6:
        return 'Power play'
    elif over <=15:
        return 'Middle'
    else:
        return 'Death'
df_deliveries['match_phase']=df_deliveries['over'].apply(get_phase)
```
